```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 1.0,0.01)
s = np.sin(2*np.pi*t)
plt.rcParams['lines.color'] = 'b'
plt.plot(t,s,c='r')

c = np.cos(2*np.pi*t)
plt.rcParams['lines.linewidth'] = '8'
plt.plot(t,c)

plt.show()
```
